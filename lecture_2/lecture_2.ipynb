# Histogram Equalization
- Histogram equalization solves the problem of low-contrast images by faltenning the histogram of intensities

<img src="hist_eq.png" height="300px"/>

# How does it work?
1. Let $f$ denote the unequalized image of size $m * n$
2. Let $g$ denote the unequalized image
3. Define the probability density function of pixel intensities of $f$ to be $$p_f(k) = \frac{H(k)}{m * n}$$ where $H(k)$ denote the number of pixels having value $k$
4. Let the cumulative density function $$CDF_f(k) = Pr\{f\leq k\} = \sum_{i=0}^{k}p_f(i)$$
5. By definition $CDF_f(L-1)=1$

# Goal
## transform image $f$ to $g$ such that $p_g(x)=1$ (g has a uniform distribution)
1. Consider the transform $$g(x,y)=CDF_f[f(x,y)]$$
2. Now, we want to compute $CDF_g(x)$
$$CDF_g(x)=Pr\{g\leq x\}=Pr\{CDF_f[f]\leq x\}=Pr\{f\leq CDF_f^{-1}[x]\}$$$$=CDF_f[CDF_f^{-1}[x]]=x$$
3. By definition
$$p_g(x)=\frac{d(CDF_g(x))}{dx}=\frac{dx}{dx}=1$$
### Therefore, we proved that the aforementioned transform yields a uniform distribution

# But why is the histogram NON-FLAT?
- In the continuous domain, there is an infinite number of intensities in any interval
- In digital images, the histogram is discrete which means there is a fixed number of values in a range
- When this range is stretched, the number of values in it is maintained
- The discrete equalization process merely remaps one intensity value to another; it does not redistribute the intensity values
- This means pixels which have equal intensities will still have equal intensities after the process

# Histogram Matching
- The idea of histogram equalization (mapping original histogram distribution to the uniform distribution) can be generalized to any kind of distribution
- Consider $T$ to be the target image, and $R$ to be the reference image to be mapped to, and $M$ to be the output image
- We compute the $CDF$ for both $T$ and $R$, which we will denote as $CDF_T(I)$ and $CDF_R(I)$
- For each intensity level $I_T$ in $T$, we search for the corresponding intensity level $I_R$ in $R$ such that $CDF_T(I_T) = CDF_R(I_R)$
- Then the matching function will be $M(I_T)=I_R$

```python
def match_histogram(img, target):
    out = np.zeros_like(img)
    H,W = out.shape[:2]
    t_H, t_W = target.shape[:2]
    for ch in range(3):
        img_unique, img_indices, img_hist = np.unique(img[:,:,ch].ravel(), return_inverse=1,return_counts=1)
        target_unique, target_hist = np.unique(target[:,:,ch].ravel(), return_counts=1)
        img_hist = np.cumsum(img_hist)/(H*W)
        target_hist = np.cumsum(target_hist)/(t_H*t_W)
        img_hist = np.interp(img_hist, target_hist, target_unique)
        print(img_hist.shape)
        out[:,:,ch] = img_hist[img_indices].reshape((H,W))

    return out
```

```python
from skimage.exposure import match_histograms

fig = plt.figure(figsize=(10, 10))
target = cv2.imread('water_target.jpg')
ref = cv2.imread('water_ref.png')
images = [target, ref]
titles = ['target', 'reference', 'matched']
target_hsv = cv2.cvtColor(target, cv2.COLOR_BGR2HSV)
ref_hsv = cv2.cvtColor(ref, cv2.COLOR_BGR2HSV)

matched = match_histogram(target_hsv, ref_hsv)#match_histograms(target_hsv, ref_hsv, multichannel=True)
matched_bgr = cv2.cvtColor(matched, cv2.COLOR_HSV2BGR)

images.append(matched_bgr)

for i, img in enumerate(images):
    fig.add_subplot(1, 3, i+1)
    plt.title(titles[i])
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

# Filtering
## Noise
- Image noise is random variation of brightness or color information in images, and is usually an aspect of electronic noise
- It can be produced by the image sensor and circuitry of a scanner or digital camera
- Noise can be modeled as a random variable
### Types of noise

- #### Gaussian Noise
    - Principal sources of Gaussian noise in digital images arise during acquisition e.g. sensor noise caused by poor illumination and/or high temperature, and/or transmission e.g. electronic circuit noise.
    - Noise can be modeled as a Gaussian PDF
    - can be removed by Average Filter, or Gaussian Filter

- #### Salt-and-Pepper Noise
    - It is also known as impulse noise
    - This noise can be caused by sharp and sudden disturbances in the image signal
    - It presents itself as sparsely occurring white and black pixels
    - Can be removed by Median Filter or Morphological operations

# Non-Linear Filters
## Median Filter
- The filter sorts pixel intensities in the neighborhood, and chooses the median value
- An important property of this filter that it is edge preserving

## Bilateral Filter
- Replaces each pixel with a weighted average of nearby pixels
- It is also edge preserving
$$BF[I]_p=\frac{1}{W_p}\sum_{q\in S}G_{\sigma_p}(\left\|p-q\right\|)G_{\sigma_r}(I_p-I_q)I_q$$
where $$W_p=\sum_{q\in S}G_{\sigma_p}(\left\|p-q\right\|)G_{\sigma_r}(I_p-I_q)$$
- The function $G_{\sigma_p}$ is usually a gaussian
- $\sigma_p$ controls the spatial filter
- $\sigma_r$ controls the intensity values of pixels with big difference than the $p$ pixel

# Edge Detection
- In continuous functions, points with the largest change are found by computing the drivative of functions
- Edges are parts of image where a sharp change of intensity appear
- For any function $f(x)$, $f(x+\Delta x)$ can be exapnded around $x$ using `Taylor Series`
$$f(x+\Delta x)=f(x)+\Delta x \frac{\partial f}{\partial x}+\frac{(\Delta x)^2}{2!} \frac{\partial^2 f}{\partial x^2}+\frac{(\Delta x)^3}{3!} \frac{\partial^3 f}{\partial x^3}+...$$
$$=\sum_{n=0}^{\infty}\frac{(\Delta x)^n}{n!}\frac{\partial^n f}{\partial x^n}$$
- for finite 2d images, let $\Delta x = 1$, therefore
$$f(x+1)=\sum_{n=0}^{\infty}\frac{1}{n!}\frac{\partial^n f}{\partial x^n}$$
- let $\Delta x = -1$, therefore
$$f(x-1)=\sum_{n=0}^{\infty}\frac{(-1)^n}{n!}\frac{\partial^n f}{\partial x^n}$$
- Using those 2 equations, and approximatinng to the first order
$$\frac{\partial f}{\partial x} \approx \frac{f(x+1)-f(x-1)}{2}$$
- And second derivatives (`Laplacian`)
$$\frac{\partial^2 f}{\partial x^2} \approx f(x+1)-2f(x)+f(x-1)$$

# This image illustrates this concept

<img src="ramp.png" auto/>

### The summary is
-  First-order derivatives generally produce thicker edges
- Second-order derivatives have a stronger response to fine detail, such as thin lines, isolated points, and noise
- Second-order derivatives produce a double-edge response at ramp and step transitions in intensity
-  The sign of the second derivative can be used to determine whether a transition into an edge is from light to dark or dark to light.

## Laplacian edge detection problem
- Due to the nature of laplacian that it is more aggressive to intensity changes, if an edge is thick enough, it produces an edge at the start and end of an edge (double-edge effect)
- Laplacian produces negative and positive values
- A simpe approach to handle this is to take the absolute values, but it results in thicker edges
- A better solution is to take positive values only, usually above a certain threshold to avoid noise
- Double edges can be handled by defining the size of laplacian to be biger than the desired edge

## Sobel Operator

<img src="sobel.png" auto/>

- Sobel Operator can be used to detect horizontal and vertical edges using 2 kernels
- Then, the 2 images are combined into one image in one of 2 ways
    - using the $L_1$ norm, also called city-block norm, or manhattan norm $$G = |G_{x}|+|G_{y}|$$
    - using the $L_2$ norm, or the Euclidian norm $$G = \sqrt{G_{x}^{2}+G_{y}^{2}}$$
- Then, we can compute the gradient direction using trigonometry 
$$\Theta = arctan\bigl(\frac{G_{y}}{G_{x}}\bigr)$$


```python
img = imread('madara.jpg', as_gray=True)
plt.figure(figsize=(10, 10))
plt.imshow(img)
```

```python
Gy = sobel(img, axis=0)
Gx = sobel(img, axis=1)

plt.figure(figsize=(15, 15))
plt.subplot(211)
plt.imshow(Gx)
plt.title('Derivative in x direction')
plt.axis('off')

plt.subplot(212)
plt.imshow(Gy)
plt.title('Derivative in y direction')
plt.axis('off')

plt.show()
```

```python
G = np.sqrt(Gx**2+Gy**2)
plt.figure(figsize=(15, 15))
plt.imshow(G)
plt.title('Edges')
plt.axis('off')
plt.show()
```
