# Canny Edge Detection
So far, we have discussed one technique of edge detection, which is by convolving an edge detector around the image. This technique is quite useful; however, sometimes it might produce some false positive edges because of noise, for example.

The algorithm can be divided into 5 main parts:

1. Smoothing (already discussed)
2. Calculating Image gradients (already discussed)
3. Non-maximum supression
4. Double thresholding
5. Edge linking

```python
img = imread('iguana.png', as_gray=True)*255
img = img.astype('uint8')

img = cv2.medianBlur(img, 5)/255.0

G, theta = gradient(img)

plt.figure(figsize=(15, 15))

plt.subplot(211)
plt.imshow(img)
plt.title('Filtered image')
plt.axis('off')

plt.subplot(212)
plt.imshow(G)
plt.title('Edges')
plt.axis('off')

plt.show()
```

# 3. Non-maximum suppression
- Edges extracted from the gradient of the smoothed image are quite thick and blurry
- This step is to convert the "blurred" edges into "sharp" edges
- This is done by preserving all local maxima in the gradient image and discarding everything else

## For each pixel (x,y) in the gradient image:
1. Round the gradient direction $\Theta[y,x]$ to the nearest 45 degrees, corresponding to the use of an 8-connected neighbourhood
2. Compare the edge strength of the current pixel with the edge strength of the pixel in the positive and negative gradient directions. For example, if the gradient direction is south (theta=90), compare with the pixels to the north and south
3. If the edge strength of the current pixel is the largest, preserve the value of the edge strength. If not, remove the value

```python
nms = non_maximum_suppression(G, theta)
print(np.max(nms))
plt.imshow(nms)
plt.title('Non-maximum suppressed')
plt.axis('off')
plt.show()
```

## Double Thresholding

- The edge-pixels remaining after the non-maximum suppression step are (still) marked with their strength pixel-by-pixel
- Many of these will probably be true edges in the image, but some may be caused by noise or color variations, for instance, due to rough surfaces
- The simplest way to differentiate between these would be to use a threshold, so that only edges stronger that a certain value would be preserved
- The Canny edge detection algorithm uses double thresholding as follows:

    - Edge pixels stronger than the high threshold are marked as strong
    - Edge pixels weaker than the low threshold are suppressed
    - Edge pixels between the two thresholds are marked as weak.


```python
low_threshold = 0.02
high_threshold = 0.03

strong_edges, weak_edges = double_thresholding(nms, high_threshold, low_threshold)
assert(np.sum(strong_edges & weak_edges) == 0)

edges=strong_edges * 1.0 + weak_edges * 0.5

plt.figure(figsize=(14, 6))
plt.subplot(121)
plt.imshow(strong_edges)
plt.title('Strong Edges')
plt.axis('off')

plt.subplot(122)
plt.imshow(edges)
plt.title('Strong+Weak Edges')
plt.axis('off')

plt.show()
```

## Edge Linking

- Strong edges are interpreted as “certain edges”, and can immediately be included in the final edge image
- Weak edges are included if and only if they are connected to strong edges”
- The weak edges can either be due to true edges or noise/color variations
- The latter type will probably be distributed independently of edges on the entire image, and thus only a small amount will be located adjacent to strong edges
- Weak edges due to true edges are much more likely to be connected directly to strong edges.

```python
edges = link_edges(strong_edges, weak_edges)

plt.imshow(edges)
plt.axis('off')
plt.show()
```

# Automatic Canny

```python
def auto_canny(image, sigma=0.1):
    v = np.median(image)
    lower = int(max(0, (1 - sigma) * v))
    upper = int(min(255, (1.0 + sigma) * v))
    print(f'lower {lower}, upper {upper}')
    return cv2.Canny(image, lower, upper, L2gradient=True)
```

```python
sigmas = [0.1, 0.5, 2]

plt.figure(figsize=(20, 5))
for i in range(len(sigmas)):
    plt.subplot(1,3,i+1)
    plt.imshow(auto_canny(img_cv, sigmas[i]))
    plt.title(f'sigma={sigmas[i]}')
    plt.axis('off')
plt.show()
```

## Graph Traversal Algorithms

<img src="graph.jpg" auto align="right"/>

- A graph is a mathematical structure that represent the relation between nodes (vertices) vis edges
- An image can be represented as a graph as follows:
    - Each pixel is a node in the graph
    - Nodes are connected by edges that represent 4-connectivity or 8-connectivity
- A graph can be traversed in 2 ways
### BFS (Breadth-First Search)
The algorithm starts at the root node and explores as far as possible along each branch before backtracking.
### DFS (Depth-First search)
It starts at the tree root, and explores all of the neighbor nodes at the present depth prior to moving on to the nodes at the next depth level

## A simple algorithm to find a global threshold automatically

1. Choose an initial threshold value $T$ (the mean of the histogram is a good choice)
2. Segment the image into 2 groups, and compute the mean intensity of each group $\mu_1, \mu_2$ respectively
3. Compute a new threshold value $$T=\frac{1}{2}(\mu_1+\mu_2)$$
4. Repeat steps 2 and 3 until $T$ converges

## Problems
- The final threshold can change drastically depending on the initial threshold

## Otsu's Method
- The method is optimum in the sense that it minimizes the in-class (intra-class) variance, hence, increases the between-class (inter-class) variance
- The basic idea is that a threshold giving the best separation between classes in terms of their intensity values would be the best (optimum) threshold
- Otsu’s method has the important property that it is based entirely on computations performed on the histogram of an image, which is computationally inexpensive

## Algorithm

1. Compute the histogram of the image normalized so that it represents the PDF of image intensities
2. Compute the expectation (mean) of the PDF
$$\mu_G=\sum_{i=0}^{L-1}i*p(i)$$ where L represents the number of intensity levels (256 in gray scale images)
3. For $K$ class segmentation, we need to compute $K-1$ threshold values
4. Compute the between-class (inter-class) variance which is defined as
$$\sigma_B^2=\sum_{i=1}^{K}P_k*(\mu_k-\mu_G)^2$$
where $P_1=\sum_{i=0}^{k_1}p(i)$, $P_2=\sum_{i=k_1+1}^{k_2}p(i)$ and so on

and $\mu_1=\frac{1}{P_1}\sum_{i=0}^{k_1}i*p(i)$, $\mu_2=\frac{1}{P_2}\sum_{i=k_1+1}^{k_2}i*p(i)$ and so on

5. Choose $k_1, k_2, ..., k_{K-1}$ such that it maximizes $\sigma_B^2$
$$\underset{0<k_1<k_2<...<k_{K-1}<L-1}{\operatorname{argmax}}\sigma_B^2(k_1, k_2, ..., k_{K-1})$$

```python
img = cv2.imread('otsu_test.tif', 0)
ret1, thres_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)
ret2, thres_binary = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV)

plt.figure(figsize=(12, 12))

plt.subplot(221)
plt.title(f'Original Image')
plt.imshow(img)
plt.axis('off')


plt.subplot(222)
plt.title('Histogram')
plt.hist(img.flat, 256, (0, 256), 1)

plt.subplot(223)
plt.title(f'Thresholded OTSU k={ret1}')
plt.imshow(thres_otsu)
plt.axis('off')

plt.subplot(224)
plt.title(f'Thresholded BINARY k={ret2}')
plt.imshow(thres_binary)
plt.axis('off')

plt.show()
```

# Using Edge values in Thresholding
- One approach for improving the shape of histograms is to consider only those pixels that lie on or near the edges between objects and the background.
- An immediate and obvious improvement is that histograms should be less dependent on the relative sizes of objects and background
- For instance, the histogram of an image composed of a small object on a large background area (or vice versa) would be dominated by a large peak because of the high concentration of one type of pixels
- Using pixels that satisfy some simple measures based on gradient and Laplacian operators has a tendency to deepen the valley between histogram peaks

```python
img = cv2.imread('cells.tif', 0)
ret, thres = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)

plt.figure(figsize=(20, 5))

plt.subplot(131)
plt.title(f'Original Image')
plt.imshow(img)
plt.axis('off')


plt.subplot(132)
plt.title('Histogram')
plt.hist(img.flat, 256, (0, 256), 1)

plt.subplot(133)
plt.title(f'Thresholded')
plt.imshow(thres)
plt.axis('off')

plt.show()
```

# Adaptive Thresholding
- The problem with ordinary thresholding that it applies a global threshold value to all the pixels of the image. That requires perfect photo with perfect lighting conditions.
- However, due to change of lighting conditions in one image, it is difficult to get good results by applying the same threshold value to all pixels without taking into account the neighbouring pixels.
- That's why we have `Adaptive Thresholding`.
- It automatically calculates the threshold value for every block of the image.

## Let's consider this photo

![title](sudoku.jpg)

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 13))
image = cv2.imread('sudoku.jpg', 0)
th, output = cv2.threshold(image, 100,255,cv2.THRESH_BINARY_INV)
adaptive = cv2.adaptiveThreshold(cv2.GaussianBlur(image, (5, 5), 0),255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,11,2.5)

axes[0].imshow(output, cmap='gray')
axes[0].set_title('Global Thresholding', fontsize=20)
axes[1].imshow(adaptive, cmap='gray')
axes[1].set_title('Adaptive Thresholding', fontsize=20)
```

# Clustering Techniques
- Clustering is a form of non-supervised machine learning, since there is no data labeling happening before the training process
- Clustering is the task of dividing  data points into a number of groups such that data points in the same groups are more similar to other data points in the same group and dissimilar to the data points in other groups.
- It is basically a collection of objects on the basis of similarity and dissimilarity between them.

# K-Means
- A clustering algorithm where we have k-center points to wich we assign other points with the goal of minimizing variance in each cluster
- Let $c$ be the vector of cluster centers and $\delta$ the membership array of every data point, the goal is to minimize the sum of square distances of each point (SSD)
$$c^*, \delta^*=\underset{c, \delta}{\operatorname{argmin}}\frac{1}{N}\sum_{i}^{N}\sum_{j}^{K}\delta_{ij}(x_i-c_j)^2$$

## Algorithm

1. Initialize $k$ cluster centers $c_i$ by random sampling from data points
2. Randomly assign data points to the $k$ classes
3. at each iteration:
    - Assign each data point to the nearest cluster using a distance measure (Euclidian for example)
    - Update cluster centers by taking the mean of every class
    - Repeat until cluster centers converge

## Application in Computer Vision
- The algorithm can be used to cluster each pixel based on its color features, location, or even both

```python
from sklearn.cluster import KMeans

img = imread('madara.jpg')
k = 6
clf = KMeans(k)
'''
clf.predict()
clf.fit()
clf.fit_predict()
'''
dataset = img.reshape(-1, 3)
predictions = clf.fit_predict(dataset)
out = predictions.reshape(img.shape[:2])
'''
colored = np.zeros_like(img)
for i in range(k):
    colored[out == i] = np.mean(img[out == i], axis=0)
'''
fig = plt.figure()


r = k//2 + k%2
for i in range(k):
    segment = img.copy()
    segment[out != i] = [0, 0, 0]
    plt.subplot(r, 2, i+1)
    plt.axis('off')
    plt.imshow(segment)
plt.show()
```

# Feature Vector
- Feature vector does not have to be `RGB` values
- It can be any meaningful representation that distinguish image regions from each other
- Other choices such as `LAB` space, `LAB + spatial separation` are completely valid

```python
from skimage.color import rgb2lab
img = imread('madara.jpg')
img = cv2.GaussianBlur(img, (5,5), 1)
lab = rgb2lab(img)
lab /= np.max(lab, axis=(0,1))
k = 4
clf = KMeans(k)
feature = lab.copy()


xx,yy = np.meshgrid(np.linspace(0,1,img.shape[1]), np.linspace(0,1,img.shape[0]))
grid = np.dstack((xx,yy))
feature = np.dstack((grid, feature))

pred = clf.fit_predict(feature.reshape(-1, feature.shape[-1]))
out = pred.reshape(img.shape[:2])

plt.figure(figsize=(10, 10))

r = k//2 + k%2
for i in range(k):
    segment = img.copy()
    segment[out != i] = [0, 0, 0]
    plt.subplot(r, 2, i+1)
    plt.axis('off')
    plt.imshow(segment)
plt.show()
```
