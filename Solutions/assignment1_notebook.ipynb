# In this task, you're asked to write a program that does the following
- Read video from camera
- crop a 400*400 box
- draw a blue 20*20 square in the upper left corner
- draw a red 20*20 square in the upper right corner

```python
import cv2
cap = cv2.VideoCapture(0)
while 1:
    _, frame = cap.read()
    frame = frame[:400,:400]
    frame = draw_solid_square_fast(frame, 0, 0, 20, (255,0,0))
    frame = draw_solid_square_fast(frame, 400-20, 0, 20, (0,0,255))
    cv2.imshow('frame', frame)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
